```python
import wave
import audioop

def remove_low_volume(input_file, output_file, threshold_db=20):
    # 입력 오디오 파일 열기
    with wave.open(input_file, 'rb') as wave_file:
        # 오디오 파일의 속성 얻기
        params = wave_file.getparams()

        # 읽은 프레임 데이터
        frames = wave_file.readframes(params.nframes)

        # 바이트 순서 바꾸기
        frames_swapped = audioop.byteswap(frames, params.sampwidth)

        # 음성 데이터를 리스트로 변환
        audio_data = list(frames_swapped)
        print('original: ', audio_data[:30])

        # 특정 데시벨 이하의 값은 0으로 만듦
        threshold_linear = 10 ** (threshold_db / 20.0)
        print("threshold: ", threshold_linear)
        audio_data = [0 if abs(sample) < threshold_linear else sample for sample in audio_data]
        print(audio_data[:30])

    # 출력 오디오 파일 생성
    with wave.open(output_file, 'wb') as output_wave:
        # 출력 파일의 속성 설정
        output_wave.setparams(params)

        # 리스트를 문자열로 변환하여 저장
        output_wave.writeframes(audioop.byteswap(bytes(audio_data), params.sampwidth))
```

```python
# 사용 예시
input_file_path = "/content/chickenangel.wav"
output_file_path = "remove_chickenangel.wav"
remove_low_volume(input_file_path, output_file_path)
```
