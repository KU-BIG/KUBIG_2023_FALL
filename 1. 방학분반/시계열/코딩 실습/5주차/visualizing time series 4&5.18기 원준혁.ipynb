```python
import pandas as pd
df=pd.read_csv("walmart_cleaned.csv")
```

```python
df.groupby(["Date"]).count()
```

```python
cond1=df["Dept"]==1
df1=df.loc[cond1]
```

```python
df1.groupby(["Store"]).count()
```

```python
cond2=df1["Store"]==43
df2=df1.loc[cond2]
```

```python
df2
```

```python
df_new=pd.DataFrame()
df_new["Date"]=df2["Date"]
df_new["Weekly_Sales"]=df2["Weekly_Sales"]
df_new["Temperature"]=df2["Temperature"]
df_new["Fuel_Price"]=df2["Fuel_Price"]
df_new["CPI"]=df2["CPI"]
```

```python
df_new.reset_index(drop=True)
```

```python
df_new["Date"]=pd.to_datetime(df_new['Date'])
df_new=df_new.set_index('Date')
```

```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
ax=df_new.plot()
plt.show()
```

```python
ax=df_new.plot(colormap='Dark2')
df_new_summary=df_new.describe()

ax.table(cellText=df_new_summary.values,colWidths=[0.3]*len(df_new.columns),
         rowLabels=df_new_summary.index,colLabels=df_new_summary.columns,loc='top')

plt.show()
```

```python
df_new.plot(subplots=True,linewidth=1, layout=(2,2), figsize=(16,10), sharex=False, sharey=False)
```

```python
import statsmodels.api as sm
my_dict={}
ts_names=df_new.columns
for i in ts_names :
  ts_decom=sm.tsa.seasonal_decompose(df_new[i])
  my_dict[i]=ts_decom
```

```python
my_dict_trend={}
for i in ts_names :
  my_dict_trend[i]=my_dict[i].trend

trend_df=pd.DataFrame.from_dict(my_dict_trend)
```

```python
trend_corr=trend_df.corr(method='spearman')
```

```python
trend_corr
```

```python
fig=sns.clustermap(trend_corr, annot=True, linewidths=1)
plt.setp(fig.ax_heatmap.yaxis.get_majorticklabels(),rotation=0)
plt.setp(fig.ax_heatmap.xaxis.get_majorticklabels(),rotation=90)
```
